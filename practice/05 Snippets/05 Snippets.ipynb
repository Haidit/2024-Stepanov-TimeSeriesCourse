```python
def get_label(labels_3d):
    labels = [0 for _ in range(len(labels_3d))]
    for i in range(len(labels_3d)):
        counter = {}
        for label in labels_3d[i]:
            if label in counter:
                counter[label] += 1
            else:
                counter[label] = 1
        labels[i] = max(counter, key=counter.get)
    return labels
```
